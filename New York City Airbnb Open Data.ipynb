## Data Cleaning On New York City Airbnb Open Data

##### This repository focuses on cleaning the New York City Airbnb Open Data to prepare it for insightful analysis. 
##### The cleaning process includes several key steps to ensure data accuracy, consistency, and reliability.

### Importing Libraries

```python
import pandas as pd
```

### Loading the Dataset

```python
df = pd.read_csv('/Users/daniellaiwagoh/Downloads/AB_NYC_2019.csv')
df
```

### Removing Duplicate Records
#### To maintain data uniqueness, duplicate records are identified and removed using the `drop_duplicates` function.


```python
df.drop_duplicates()
```

```python
df=df.drop_duplicates()
df
```

### Handling Missing Data
#### Missing values in the dataset are addressed by filling them with appropriate placeholders. 
#### The `last_review` column is filled with `'No Review'` for missing values, and `reviews_per_month` is set to `0` where data is unavailable.


```python
df['last_review'] = df['last_review'].fillna('No Review') 
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df
```

### Duplicate Removal by ID
#### To ensure data uniqueness, duplicate records are removed based on the `id` column, retaining only the first occurrence.


```python
df = df.drop_duplicates(subset=['id'], keep='first')
df
```

### Removing Invalid Values
#### To ensure data accuracy, records with negative values in the `price` and `availability_365` columns are removed, keeping only valid non-negative entries.


```python
df = df[(df['price'] >= 0) & (df['availability_365'] >= 0)]
df
```

### Outlier Detection and Removal
#### To prevent skewed analysis, outliers in the `price` and `minimum_nights` columns are removed by capping values at the 99th percentile.


```python
df = df[df['price'] <= df['price'].quantile(0.99)]
df = df[df['minimum_nights'] <= df['minimum_nights'].quantile(0.99)]
df
```

### Standardizing Text Data
#### To ensure consistency across the dataset, the `name` and `neighbourhood` columns are standardized to title case for improved readability and uniformity.


```python
df['name'] = df['name'].str.title()  
df['neighbourhood'] = df['neighbourhood'].str.title()
df
```

### Filtering Data by Geographic Boundaries
#### To ensure location accuracy, the dataset is filtered to retain only records within valid latitude and longitude boundaries corresponding to New York City.


```python
df = df[(df['latitude'].between(40.4774, 40.9176)) & (df['longitude'].between(-74.2591, -73.7004))]
df
```

### Standardizing Date Formats
#### The `last_review` column is converted to a standardized datetime format, ensuring consistency and handling any invalid date entries by coercing them to `NaT`.


```python
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
df
```

### Exploring Dataset Structure
#### The `df.info()` function provides an overview of the dataset, including column names, data types, non-null counts, and memory usage.


```python
df.info()
```

```python
df
```

### Standardizing Column Names
#### To improve readability and maintain consistency, all column names are converted to title case.


```python
df.columns = df.columns.str.title()
df
```

### Checking for Missing Data
#### The `isnull().sum()` function is used to identify missing values in each column of the dataset, helping to assess where data imputation or cleaning is needed.


```python
print(df.isnull().sum())
df
```

### Missing Data Handling
#### To maintain data completeness, missing values are filled with appropriate default values:

```python
df['Name'] = df['Name'].fillna('No Name')  
df['Host_Name'] = df['Host_Name'].fillna('No Host Name')  
df['Last_Review'] = df['Last_Review'].fillna('No Review')  
df['Reviews_Per_Month'] = df['Reviews_Per_Month'].fillna(0)
```

```python
df
```

### Verifying Missing Data Handling
#### After filling missing values, `df.isnull().sum()` is used to verify that there are no remaining missing entries in the dataset.


```python
print(df.isnull().sum())
```

```python
df
```

### Ensuring Correct Data Types
#### To ensure data consistency and facilitate analysis, we convert specific columns to their appropriate data types. The `pd.to_numeric()` function is used to convert numerical columns, while `pd.to_datetime()` handles date conversions. 

```python

df['Host_Id'] = pd.to_numeric(df['Host_Id'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Minimum_Nights'] = pd.to_numeric(df['Minimum_Nights'], errors='coerce')
df['Last_Review'] = pd.to_datetime(df['Last_Review'], errors='coerce')
```

```python
print(df.dtypes)
```

### Handling Missing Values
#### To ensure data completeness and accuracy, missing values in critical columns are filled using appropriate strategies 

```python
df['Reviews_Per_Month'] = df['Reviews_Per_Month'].fillna(df['Reviews_Per_Month'].median())
df['Last_Review'] = df['Last_Review'].fillna(df['Last_Review'].min())
```

### Standardizing Categorical Data

#### To ensure consistency and readability, categorical columns are standardized by:

#### - Removing leading and trailing whitespace.
#### - Converting text to title case for uniform formatting.

```python
df['Room_Type'] = df['Room_Type'].str.strip().str.title()
df['Neighbourhood_Group'] = df['Neighbourhood_Group'].str.strip().str.title()
df
```

### Filtering Data Based on Price Range

#### To maintain relevant data and eliminate extreme outliers, the dataset is filtered to include only properties with a price between 10 and 1000.

```python
df = df[(df['Price'] > 10) & (df['Price'] < 1000)]
df
```

### Removing Duplicate Records

#### To ensure that the dataset contains only unique entries, duplicates based on the combination of `Id`, `Host_Id`, and `Name` were removed, keeping only the first occurrence.


```python
df = df.drop_duplicates(subset=['Id', 'Host_Id', 'Name'], keep='first')
df
```
